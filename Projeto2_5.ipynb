```python
#Lendo o data set com os nutrientes do cardapio de alguns fastfoods
df = pd.read_csv("FastFoodNutritionMenuV3 - FastFoodNutritionMenuV3.csv")
df.head()
```

```python
new_columns = [column.replace('\n', ' ') for column in df.columns]
df = df.rename(columns=dict(zip(df.columns, new_columns)))

df.columns
```

# Limpando o dataset

```python
#conversão das colunas do DataFrame para o tipo de dado numérico
df["Calories"] = pd.to_numeric(df["Calories"], errors='coerce')
df["Calories from Fat"]  = pd.to_numeric(df["Calories from Fat"] , errors='coerce')
df["Total Fat (g)"]  = pd.to_numeric(df["Total Fat (g)"] , errors='coerce')
df["Saturated Fat (g)"]  = pd.to_numeric(df["Saturated Fat (g)"] , errors='coerce')
df["Trans Fat (g)"]  = pd.to_numeric(df["Trans Fat (g)"] , errors='coerce')
df["Cholesterol (mg)"]  = pd.to_numeric(df["Cholesterol (mg)"] , errors='coerce')
df["Sodium  (mg)"]  = pd.to_numeric(df["Sodium  (mg)"] , errors='coerce')
df["Carbs (g)"]  = pd.to_numeric(df["Carbs (g)"] , errors='coerce')
df["Fiber (g)"]  = pd.to_numeric(df["Fiber (g)"] , errors='coerce')
df["Sugars (g)"]  = pd.to_numeric(df["Sugars (g)"] , errors='coerce')
df["Protein (g)"]  = pd.to_numeric(df["Protein (g)"] , errors='coerce')
```

### **ANALISE EXPLORATORIA COM MEDIDAS E GRÁFICOS**

```python
# Média das caracteristicas dos items de cada compania
dfMeanByCompany = df.groupby(['Company']).agg({
  'Calories': "mean",
  'Calories from Fat': "mean",
  'Total Fat (g)': "mean",
  'Saturated Fat (g)': "mean",
  'Trans Fat (g)': "mean",
  'Cholesterol (mg)': "mean",
  'Sodium  (mg)': "mean",
  'Carbs (g)': "mean",
  'Fiber (g)': "mean",
  'Sugars (g)': "mean",
  'Protein (g)': "mean"
})

dfMeanByCompany
```

```python
#Mediana das caracteristicas dos Items de cada compania
dfMedianByCompany = df.groupby(['Company']).agg({
  'Calories': "median",
  'Calories from Fat': "median",
  'Total Fat (g)': "median",
  'Saturated Fat (g)': "median",
  'Trans Fat (g)': "median",
  'Cholesterol (mg)': "median",
  'Sodium  (mg)': "median",
  'Carbs (g)': "median",
  'Fiber (g)': "median",
  'Sugars (g)': "median",
  'Protein (g)': "median"
})

dfMedianByCompany
```

```python
# Calcula o maior valor de gordura saturada para cada empresa
max_saturated_fat = df.groupby("Company")["Saturated Fat (g)"].max()
# Filtra o DataFrame original para manter apenas as linhas correspondentes aos maiores valores de gordura saturada
df_filtered = df.loc[df.groupby("Company")["Saturated Fat (g)"].idxmax()]
df_filtered = df_filtered.sort_values(by="Saturated Fat (g)", ascending=False)
#define as colunas para visualização dos itens
df_filtered_gordura = df_filtered[['Company', 'Item','Saturated Fat (g)']]

# Calcula o maior valor de colesterol para cada empresa
max_cholesterol = df.groupby("Company")["Cholesterol (mg)"].max()
# Filtra o DataFrame original para manter apenas as linhas correspondentes aos maiores valores de colesterol
df_filtered_cholesterol = df.loc[df.groupby("Company")["Cholesterol (mg)"].idxmax()]
df_filtered_cholesterol = df_filtered_cholesterol.sort_values(by="Cholesterol (mg)", ascending=False)
# Define as colunas para visualização dos itens
df_filtered_cholesterol = df_filtered_cholesterol[['Company', 'Item','Cholesterol (mg)']]

# Define os eixos dos gráficos para colesterol
x_cholesterol = df_filtered_cholesterol["Company"]
y_cholesterol = df_filtered_cholesterol["Cholesterol (mg)"]

#define os eixos dos gráficos
x = df_filtered["Company"]
y = df_filtered["Saturated Fat (g)"]

# Criando uma grade de subplots
plt.figure(figsize=(12, 6))

# Número de barras
num_bars = len(x)

# Gerando as cores dos graficos
cores_gordura = plt.cm.coolwarm(np.linspace(0, 1, num_bars))
cores_colesterol = plt.cm.twilight_shifted(np.linspace(0, 1, num_bars))

# Gráfico para gordura saturada
plt.subplot(1, 2, 1)  # subplot de 1 linha e 2 colunas, posição 1
plt.bar(x, y, linewidth=0.6, color=cores_gordura)
plt.xlabel("Empresa")
plt.ylabel("Gordura Saturada (g)")
plt.title("Restaurante com o item de maior Gordura Saturada")
plt.xticks(rotation=45)
for i, v in enumerate(y):
    plt.text(i, v + 0.1, str(v), ha='center')

# Gráfico para colesterol
plt.subplot(1, 2, 2)  # subplot de 1 linha e 2 colunas, posição 2
plt.bar(x_cholesterol, y_cholesterol, linewidth=0.6, color=cores_colesterol)
plt.xlabel("Empresa")
plt.ylabel("Colesterol (mg)")
plt.title("Restaurante com o item de maior Colesterol")
plt.xticks(rotation=45)
for i, v in enumerate(y_cholesterol):
    plt.text(i, v + 0.1, str(v), ha='center')

plt.tight_layout()  # Ajusta o layout dos subplots para evitar sobreposição
plt.show()
# Mostrando os DataFrames
display(df_filtered_gordura)
display(df_filtered_cholesterol)
```

```python
# Calcula o maior valor de sódio para cada empresa
max_sodium = df.groupby("Company")["Sodium  (mg)"].max()
# Filtra o DataFrame original para manter apenas as linhas correspondentes aos maiores valores de sódio
df_filtered_sodium = df.loc[df.groupby("Company")["Sodium  (mg)"].idxmax()]
df_filtered_sodium = df_filtered_sodium.sort_values(by="Sodium  (mg)", ascending=False)
# Define as colunas para visualização dos itens
df_filtered_sodium = df_filtered_sodium[['Company', 'Item','Sodium  (mg)']]

# Calcula o maior valor de açúcares para cada empresa
max_sugars = df.groupby("Company")["Sugars (g)"].max()
# Filtra o DataFrame original para manter apenas as linhas correspondentes aos maiores valores de açúcares
df_filtered_sugars = df.loc[df.groupby("Company")["Sugars (g)"].idxmax()]
df_filtered_sugars = df_filtered_sugars.sort_values(by="Sugars (g)", ascending=False)
# Define as colunas para visualização dos itens
df_filtered_sugars = df_filtered_sugars[['Company', 'Item','Sugars (g)']]

# Define os eixos dos gráficos para sódio
x_sodium = df_filtered_sodium["Company"]
y_sodium = df_filtered_sodium["Sodium  (mg)"]

# Define os eixos dos gráficos para açúcares
x_sugars = df_filtered_sugars["Company"]
y_sugars = df_filtered_sugars["Sugars (g)"]

# Criando uma grade de subplots
plt.figure(figsize=(12, 6))

# Número de barras
num_bars = len(x)

# Gerando as cores dos gráficos
cores_sodium = plt.cm.coolwarm(np.linspace(0, 1, num_bars))
cores_sugars = plt.cm.twilight_shifted(np.linspace(0, 1, num_bars))

# Gráfico para sódio
plt.subplot(1, 2, 1)  # subplot de 1 linha e 2 colunas, posição 1
plt.bar(x_sodium, y_sodium, linewidth=0.6, color=cores_sodium)
plt.xlabel("Empresa")
plt.ylabel("Sódio (mg)")
plt.title("Restaurante com o item de maior Sódio")
plt.xticks(rotation=45)
for i, v in enumerate(y_sodium):
    plt.text(i, v + 0.1, str(v), ha='center')

# Gráfico para açúcares
plt.subplot(1, 2, 2)  # subplot de 1 linha e 2 colunas, posição 2
plt.bar(x_sugars, y_sugars, linewidth=0.6, color=cores_sugars)
plt.xlabel("Empresa")
plt.ylabel("Açúcares (g)")
plt.title("Restaurante com o item de maior Açúcares")
plt.xticks(rotation=45)
for i, v in enumerate(y_sugars):
    plt.text(i, v + 0.1, str(v), ha='center')

plt.tight_layout()  # Ajusta o layout dos subplots para evitar sobreposição
plt.show()

# Mostrando os DataFrames
display(df_filtered_sodium)
display(df_filtered_sugars)
```
